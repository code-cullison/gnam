## Get Real Event Data

```python
from gnam.events.gevents import gevents as gevents
from gnam.events.gstations import gstations as gstations
from obspy import Stream
import matplotlib.pyplot as plt
import numpy as np
import pickle

# Unpickle
f = open('./model_bbox.pickle', 'rb')
gf_bbox = pickle.load(f)
f.close()
print('bbox after pickle :\n',gf_bbox)

f = open('./enorth.pickle', 'rb')
enorth = pickle.load(f)
f.close()

f = open('./gevents_test2.pickle', 'rb')
dill_snorth = pickle.load(f)
f.close()


#get streams
ekey = 3
e1_st3 = dill_snorth.getStream1(ekey,3)
e1_st4 = dill_snorth.getStream1(ekey,4)
e2_st3 = dill_snorth.getStream2(ekey,3)
e2_st4 = dill_snorth.getStream2(ekey,4)
ez_st3 = dill_snorth.getStreamZ(ekey,3)
ez_st4 = dill_snorth.getStreamZ(ekey,4)

print('len(ez_st3):',len(ez_st3))
print('len(ez_st4):',len(ez_st4))
print('len(e1_st3):',len(e1_st3))
print('len(e1_st4):',len(e1_st4))
print('len(e2_st3):',len(e2_st3))
print('len(e2_st4):',len(e2_st4))

e_st3 = e1_st3 + e2_st3 + ez_st3
e_st4 = e1_st4 + e2_st4 + ez_st4

for i in range(1):
    print('e3:\n',vars(e_st3[i+58]))
    print('e4:\n',vars(e_st4[i+58]))
te_st3 = Stream()
te_st4 = Stream()

ns = len(ez_st3)
for i in range(ns):
    te_st3 += e_st3[i+2*ns]
    te_st3 += e_st3[i+ns]
    te_st3 += e_st3[i]
    
    te_st4 += e_st4[i+2*ns]
    te_st4 += e_st4[i+ns]
    te_st4 += e_st4[i]

e_st3 = te_st3
e_st4 = te_st4
print('e3:\n',e_st3[0:3])
print('e4:\n',e_st4[0:3])

inv3 = dill_snorth.get_inventory(3)
inv4 = dill_snorth.get_inventory(4)

net3 = inv3[0]
net4 = inv4[0]
#inv.get_orientation("GR.FUR..LHE", t)
for i in range(58):
    sid = e_st3[i*3+2].stats.network +'.'+ e_st3[i*3+2].stats.station +'..'+ e_st3[i*3+2].stats.channel 
    #print('sid:',sid)
    print('net3.orientation:',net3.get_orientation(sid))

#e_st.filter('bandpass',freqmin=1.0,freqmax=18.0,corners=4)
#e_st.plot()
```

### Try to rotate via obspy

```python
re_st3 = e_st3.copy()
re_st4 = e_st4.copy()

re_st3.rotate(method="->ZNE", inventory=inv3)
re_st4.rotate(method="->ZNE", inventory=inv4)

print('re3:\n',re_st3[0:3])
print('re4:\n',re_st4[0:3])

iet = -1
for i in range(len(e_st4)):
    station = re_st4[i].stats.station 
    channel = re_st4[i].stats.channel
    if station == 'G654' and channel == 'HHZ':
        iet = i

irt = -1
for i in range(len(re_st4)):
    station = re_st4[i].stats.station 
    channel = re_st4[i].stats.channel
    if station == 'G654' and channel == 'HHZ':
        irt = i
        
print('B4/After rotation G654.FXZ:')
for i in range(e_st4[iet].count()):
    print('b4,af = %E,%E' %(e_st4[iet].data[i],re_st4[irt].data[i]))
```

### Resample data

```python
import obspy
from obspy import Stream
from obspy import Trace
from obspy import UTCDateTime

dt = 0.001
rsamp = int(1.0/dt +0.5)
print('dt:    ',dt)
print('rsamp:',rsamp)

cre_st3 = re_st3.copy()
cre_st4 = re_st4.copy()
re_st = cre_st3 + cre_st4
print('Before resmapling:\n',re_st)

re_st.resample(1000)
print('After resmapling:\n',re_st)

re_st[0].write('junk_trace_ascii.txt',format='SAC')

itr = 0
for tr in re_st:
    print('Trace[%d].stations.channel: %s.%s' %(itr,tr.stats.station,tr.stats.channel))
    itr += 1
```

### Write Rotated Observed DATA to SPECFEM ASCII 

```python
nt = 16384 ##arbitrary but is <= tr.count()
t = dt*np.arange(nt)
for tr in re_st:
    spec_pair_list = [] #time data/amplitude pairs
    
    #Example of filename: 'NL.G094.FXX.semd'
    tr_spec_chan = '.BOO.'
    comp_char = tr.stats.channel[2] 
    if comp_char == 'Z':
        tr_spec_chan = '.FXZ.'
    elif comp_char == 'E':
        tr_spec_chan = '.FXX.'
    elif comp_char == 'N':
        tr_spec_chan = '.FXY.'
    else:
        print('you suck')
        
    print('chan:',tr_spec_chan)
    tr_filename = 'NL.' + tr.stats.station + tr_spec_chan + 'semd'
    print('filename:',tr_filename)
    
    assert nt <= len(tr.data)
    for it in range(nt):  #only works if observed data is same len or longer
        spec_pair_list.append('%E   %E\n' %(t[it],tr.data[it]))

    fqpname = './OBS_ROT/' + tr_filename
    f = open(fqpname, 'w')
    f.writelines(spec_pair_list)
    f.close()
```
